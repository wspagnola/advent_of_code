```python
import os 
import numpy as np
import pandas as pd
#os.listdir()
```

```python
test_lst = []

with open("input8_test.txt") as f:
    lines = f.readlines()
    
    for l in lines:
        test_lst.append(l.strip().replace('|', '').split())
    f.close()
print(test_lst)

X = pd.DataFrame(test_lst)
X.head()
```

```python
check = pd.concat([X.iloc[0].rename('char'), X.iloc[0].apply(len).rename('char_len')], axis = 1)

check.loc[check.char_len == 2, 'char_replace'] = "cf"
check.loc[check.char_len == 4, 'char_replace'] = "bcdf"
check.loc[check.char_len == 3, "char_replace"] = "acf"
check.loc[check.char_len == 7, "char_replace"] = "abcdefg"

check.loc[check.char_len == 2, 'num_guess'] = "1"
check.loc[check.char_len == 4, "num_guess"] = "4"
check.loc[check.char_len == 7, "num_guess"] = "8"
check.loc[check.char_len == 3, "num_guess"] = "7"
check.loc[check.char_len == 6, "num_guess"] = "0, 6, 9"
check.loc[check.char_len == 5, "num_guess"] = "2, 3, 5"


check['char_str'] =check['char'].apply(lambda x: ''.join(sorted(x)))
    
check
```
