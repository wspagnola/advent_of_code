### Check Bingo Cards

```python
def check_cards(mat):
    row_check_vec = np.sum(mat, axis = 1)
    col_check_vec = np.sum(mat, axis = 0)
    check_vec = np.concatenate((row_check_vec, col_check_vec), axis = 0)
    #print(check_vec)
    if np.max(check_vec)== 5:
        #print('Bingo')
        return(int(1)) 
    else:
        return(int(0))
```

### Play Bingo Function

```python
def play_bingo(nums, cards):
    
    card_step = np.zeros(cards.shape, dtype = 'int32')
    cur_cards = card_step
    
    for round, num in enumerate(nums):
          
        print('round:', round + 1)
        print('\tnumber:', num)
        
        card_step = (cards == num)*1
        cur_cards = cur_cards + card_step

        for i, marks in enumerate(cur_cards):
            
            bingo_check = check_cards(marks)
            print('\tbingo check {}: {}'.format(i,bingo_check))
            
            if bingo_check == 1:
                print('Bingo!')
                return([round, num, cards[i], marks])
```

## Lose bingo

```python
def lose_bingo(nums, cards):
    
    n_cards = cards.shape[0]
    card_step = np.zeros(cards.shape, dtype = 'int32')
    cur_cards = card_step
    bingo_check_cards = np.zeros(  n_cards)
    
    # Call numbers
    for round, num in enumerate(nums):
          
        print('round:', round + 1)
        print('number:', num)
        
        card_step = (cards == num)*1
        cur_cards = cur_cards + card_step

        # Cycle through cards
        for i, marks in enumerate(cur_cards):
            
            bingo_check = check_cards(marks)
            bingo_check_cards[i] = bingo_check
            #print('\tbingo check {}: {}'.format(i,bingo_check))
            
        n_card_wins = np.sum(bingo_check_cards)
        print("Wins: {} Left: {} Total: {}".format(n_card_wins,   n_cards - n_card_wins, n_cards))
        if n_card_wins == (n_cards - 1):
            print(bingo_check_cards)
            last_card = np.where(bingo_check_cards == 0)[0][0] # losing card index number
        if n_card_wins == (n_cards):
            #Note: Need to Make sure this is last winning card, hence last_card index
            return([round, num, cards[last_card], cur_cards[last_card], last_card])
        
```
