```python


def random_cave_walk():
    
    step_lst = []
    step0 = "start"
    
    while(len(step_lst) < 10):

        if step0 == "start":
            step1 = np.random.choice(['A', 'b'])
            step_lst.append(step0)
            step0 = step1
        elif step0 == "A":
            step1 = np.random.choice(['c', 'b', 'end'])
            step_lst.append(step0)
            step0 = step1
        elif step0 == "b":
            step1 = np.random.choice(['A','d', 'end'])
            step_lst.append(step0)
            step0 = step1
        elif step0 == "c":
            step1 = "A"
            step_lst.append(step0)
            step0 = step1
        elif step0 == 'd':
            step1 = "b"
            step_lst.append(step0)
            step0 = step1
        else:
            step_lst.append(step0)
            step0 = step1
            return(step_lst)
    return(step_lst)

#print(step_lst)
# if step == "start":
#     np.sampl
```

```python

def random_cave_walk1():
    
    step_lst = []
    step0 = "start"
    
    while(len(step_lst) < 100):

        if step0 == "start":
            step1 = np.random.choice(['dc', 'HN', 'kj'])
            step_lst.append(step0)
            step0 = step1
            
        elif step0 == "HN":
            step1 = np.random.choice(['dc', 'kj', 'end'])
            step_lst.append(step0)
            step0 = step1
            
        elif step0 == "dc":
            step1 = np.random.choice(['kj', 'HN', 'end'])
            step_lst.append(step0)
            step0 = step1
            
        elif step0 == "kj":
            step1 = np.random.choice(['sa', 'HN', 'dc'])
            step_lst.append(step0)
            step0 = step1
            
        elif step0 == 'sa':
            step1 = "kj"
            step_lst.append(step0)
            step0 = step1
        else:
            step_lst.append(step0)
            step0 = step1
            return(step_lst)
    return(step_lst)
```

```python
def random_cave_walk3(max_iter = 10000):
    
    step_lst = ["start"]
    step0 = "start"
    
    while(len(step_lst) < max_iter):

        if step0 == "start":
            step1 = np.random.choice(['na', 'sr', 'FP'])
            step_lst.append(step1)
            step0 = step1
            
        elif step0 == "FP":
            step1 = np.random.choice(['lh', 'sr', 'na', 'end'])
            step_lst.append(step1)
            step0 = step1
            
        elif step0 == "na":
            step1 = np.random.choice(['vy', 'mi', 'FP'])
            step_lst.append(step1)
            step0 = step1
            
        elif step0 == "sr":
            step1 = np.random.choice(['KF', 'FP', 'vy', 'mi'])
            step_lst.append(step1)
            step0 = step1
            
        elif step0 == 'lh':
            step1 = np.random.choice(['vy', 'KF', 'FP', 'end'])
            step_lst.append(step1)
            step0 = step1
            
        elif step0 == 'vy':
            step1 = np.random.choice(['OO', 'mi', 'KF', 'na', 'sr', 'lh'])
            step_lst.append(step1)
            step0 = step1 
            
        elif step0 == 'mi':
            step1 = np.random.choice(['vy', 'na', 'KF', 'sr'])
            step_lst.append(step1)
            step0 = step1  
          
        elif step0 == 'KF':
            step1 = np.random.choice(['mi', 'sr', 'lh', 'vy', 'wp', 'end'])
            step_lst.append(step1)
            step0 = step1  
            
        elif step0 == 'OO':
            step1 = np.random.choice(['vy'])
            step_lst.append(step1)
            step0 = step1  
        
        elif step0 == 'wp':
            step1 = np.random.choice(['KF'])
            step_lst.append(step1)
            step0 = step1  
            
        else:
            #step_lst.append(step0)
            step0 = step1
            return(step_lst)
    return(step_lst)
```

```python
def sim_walks(nIter):
    
    walks = []

    for i in range(nIter):
        walks.append(random_cave_walk3())
        
    return(walks)
```

```python
walk_fin = sim_walks(500000)
```

```python
output = []

for i, walk in enumerate(walk_fin):

    output.append(walk) if walk not in output else []
    #print(i, walk)

output2 = []
for i in output:
    output2.append(i) if (i.count('mi') <= 1) & (i.count('sr') <= 1) & (i.count('lh') <= 1)  & (i.count('vy') <= 1)  & (i.count('wp') <= 1) & (i.count('na') <= 1) else []
print(len(output2))
```
